```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the data

```python
dataset = pd.read_csv('Salary_Data.csv')
dataset.head()
```

splitting data

```python
x=dataset.iloc[:,:-1].values   #select all column except last with all rows
y=dataset.iloc[:,-1].values   #select last column with all rows
```

Now split the dataset with train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1) #0.2 give 20% for testing and random state 1 is to preserve reproducability
```

# data preprocessing is over.....
use linear model for linear regression model.....

training model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

prediction test

```python
y_pred=regressor.predict(x_test)
```

visualize the training set results

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
```

Visualizing the test set results

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.show()
```
